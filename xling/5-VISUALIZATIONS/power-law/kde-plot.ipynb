## Visualizing results

```python
import plotly.graph_objects as go
import plotly.figure_factory as ff
```

## All data

```python
hist_data = [
    df[COEF_VAR].loc[df['self']].values,
    df[COEF_VAR].loc[~df['self']].values,
]

groups = [
    'self',
    'other',
]

colors = [
    '#FFC30B',
    '#0041C2'
]
```

```python
fig1 = ff.create_distplot(
    hist_data=hist_data,
    group_labels=groups,
    colors=colors,
    show_hist=False
)

fig1.update_layout(template='ygridoff')
```

```python
fig1.write_html(os.path.join(DIGITAL_EXCLUSIVE_PATH, 'all-data.html'))
fig1.update_xaxes(range=xaxisrange)
fig1.write_html(os.path.join(PUBLICATION_READY_PATH, 'all-data.html'))
fig1.write_image(os.path.join(PUBLICATION_READY_PATH, 'all-data..png'), scale=5)
```

### by corpus

```python
corpus_col = 'lang'
```

```python
figs = []
```

```python
for corpus in df[corpus_col].unique():
    sub = df.loc[df[corpus_col].isin([corpus])]

    hist_data = [
        sub[COEF_VAR].loc[sub['self']].values,
        sub[COEF_VAR].loc[~sub['self']].values,
    ]

    groups = [
        'self',
        'other',
    ]

    colors = [
        '#FFC30B',
        '#0041C2'
    ]

    figs += [
        ff.create_distplot(
            hist_data=hist_data,
            group_labels=groups,
            colors=colors,
            show_hist=False
        )
    ]
    # figs[-1].update_xaxes(range=[-1,1])
    figs[-1].update_layout(template='ygridoff')
```

```python
for corpus, plot in zip(df[corpus_col].unique(), figs):
    print(corpus)
    plot.show()
    plot.write_html(os.path.join(DIGITAL_EXCLUSIVE_PATH, corpus+'.html'))
    plot.update_xaxes(range=xaxisrange)
    plot.write_html(os.path.join(PUBLICATION_READY_PATH, corpus+'.html'))
    plot.write_image(os.path.join(PUBLICATION_READY_PATH, corpus+'.png'), scale=5)
```
