```python
model_script = """
model{
    b0 ~ dunif(-100, 100)
    beta ~ dunif(-100, 100)
    phi ~ dunif(-100, 100)
    gamma ~ dunif(1, 10)
    eta ~ dunif(1, 10)
    omega ~ dunif(1, 10)
    # self_b0 ~ dunif(-100, 100)
    # other_b0 ~ dunif(-100, 100)

    for (c in 1:CCONVERSATION){
        conversation_sigma[c] ~ dnorm(omega, 1)
    }

    for (p in 1:KSPEAKER){
        speaker_sigma[p] ~ dnorm(gamma, 1)
        other_sigma[p] ~ dnorm(eta, 1)
    }

    for (r in 1:ROWS){
        mu[r] <- survey_var[r] * ((is_self[r] * phi) + (is_other[r]*beta))
        sigma[r] <- conversation_sigma[conversation[r]] + ((is_self[r] * speaker_sigma[speaker[r]]) + (is_other[r] * other_sigma[speaker[r]]))
        bc[r] <- (self_b0*is_self[r]) + (other_b0*is_other[r])
        slope[r] ~ dnorm(bc[r] + mu[r], 1/pow(sigma[r],2))
    }
}
"""
f = open('model.txt', 'w')
f.write(model_script)
f.close()
```
