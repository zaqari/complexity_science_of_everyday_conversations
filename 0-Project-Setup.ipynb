## Create data folder structure

```python
import os
```

### English dataset

```python
DATA_LOC = 'english/data'
os.mkdir(DATA_LOC)
```

First, we'll make separate folders for CHAs and Candor files

```python
# chas
os.mkdir(os.path.join(DATA_LOC, 'chas'))

# candor
os.mkdir(os.path.join(DATA_LOC, 'candor'))

# Miao-fNIRS
os.mkdir(os.path.join(DATA_LOC, 'grace_data'))
```

Every time we process a cha or candor file, we need to dump it in the `server_ready` folder, which we make here.

```python
os.mkdir(os.path.join(DATA_LOC, 'server_ready'))
```

When we've run a CEDA analysis on the data, we need to dump it into a folder to hold onto each checkpoint generated during that process.

```python
os.mkdir(os.path.join(DATA_LOC, 'ckpts'))
```

And we'll then stitch these checkpoints together after doing some minimal preprocessing of the metadata, and post that edited data to the following folder.

```python
os.mkdir(os.path.join(DATA_LOC, 'lme_ready'))
```

And just for fun, for any html, interactive charts we make, we'll save those here:

```python
os.mkdir(os.path.join(DATA_LOC, 'web_vis'))
```
